```
from gensim.models import LsiModel, TfidfModel, LdaModel
from gensim.corpora import MmCorpus, Dictionary
import pandas as pd
```

```
import semantic_lexical_lir.text_analyzer as text_analyzer
from semantic_lexical_lir.bm25 import BM25
```

```
thesaurus_file='semantic_lexical_lir/data/thesaurus.csv'
stopwords_file='semantic_lexical_lir/data/stopwords.txt'
analyzer = text_analyzer.load_analyzer(thesaurus_file=thesaurus_file, stopwords_file=stopwords_file)
```

```
topics = [i for i in range (50, 450, 50)]
models = ['tfidf', 'bm25'] + ['lsi_%s'%i for i in topics] + ['lda_%s'%i for i in topics]
models
```

### Base directories where models, indexes and queries are stored. 
You may change 'models_dir' if you want to load models  you trained by yourself (since you keep filename pattern).

```
corpus_a = {
    'models_dir': 'semantic_lexical_lir/data/corpus_a/models/',
    'data_dir': 'semantic_lexical_lir/data/corpus_a/',
    'output_dir': 'corpus_a/'
}

corpus_b = {
    'models_dir': 'semantic_lexical_lir/data/corpus_b/models/',
    'data_dir': 'semantic_lexical_lir/data/corpus_b/',
    'output_dir': 'corpus_b/'
}

corpus_a['dictionary'] = Dictionary.load(corpus_a['data_dir'] + 'dictionary.dict')
corpus_b['dictionary'] = Dictionary.load(corpus_a['data_dir'] + 'dictionary.dict')

load_test_bows = lambda c: MmCorpus(c['data_dir'] + 'bows_test')
corpus_a['test_bows'] = load_test_bows(corpus_a)
corpus_b['test_bows'] = load_test_bows(corpus_b)
```

```
def load_models(corpus):
    folder = corpus['models_dir']
    models = {}
    models['tfidf'] = TfidfModel.load(folder + 'tfidf')
    models['lsi'] = {t: LsiModel.load(folder + 'lsi_%s'% t)  for t in topics}
    models['lda'] = {t: LdaModel.load(folder + 'lda_%s'% t)  for t in topics}
    models['bm25'] = BM25(corpus['test_bows'])
    return models

corpus_a['models'] = load_models(corpus_a)
corpus_b['models']  = load_models(corpus_b)

corpus_a['models']
```

```
from gensim.similarities import MatrixSimilarity

def load_index(folder):
    indexes = {}
    indexes['tfidf'] = MatrixSimilarity.load(folder + 'tfidf_index')
    indexes['lsi'] = {t: MatrixSimilarity.load(folder + 'lsi_%s_index'% t)  for t in topics}
    indexes['lda'] = {t: MatrixSimilarity.load(folder + 'lda_%s_index'% t)  for t in topics}
    return indexes

corpus_a['indexes'] = load_index(corpus_a['models_dir'])
corpus_b['indexes'] = load_index(corpus_b['models_dir'])

corpus_a['indexes']
```

```
def load_queries(queries_file):
    queries = pd.read_csv(queries_file, index_col=0)
    queries['matter'] = queries.matter.apply(eval)
    queries['citations'] = queries.citations.apply(eval)
    return queries
    
corpus_a['queries'] = load_queries(corpus_a['data_dir'] + 'queries.csv')
corpus_b['queries'] = load_queries(corpus_b['data_dir'] + 'queries.csv')
```

```
def vectorize(text, corpus, models_pipeline):
    bow = corpus['dictionary'].doc2bow(analyzer(text))
    vect = bow
    for m in models_pipeline:
        vect = m[vect]
    return vect

tfidf_vect = lambda text, corpus: vectorize(text, corpus, [corpus['models']['tfidf']]) 
bm25_vect = lambda text, corpus: vectorize(text, corpus, []) 
lsi_vect = lambda text, corpus, ntopics: vectorize(text, corpus, [corpus['models']['tfidf'], corpus['models']['lsi'][ntopics]])
lda_vect = lambda text, corpus, ntopics: vectorize(text, corpus, [corpus['models']['lda'][ntopics]])
```

```
# query an index from a corpus with a model.
def query_index(text, corpus, model, fn_vect, ntopics=None):
    sims = []
    if ntopics:
        sims = corpus['indexes'][model][ntopics][fn_vect(text, corpus, ntopics)]          
    else:
        if model == 'bm25':
            sims = corpus['models']['bm25'].similarities(fn_vect(text, corpus))
            sims = [s[1] for s in sims]
        else:
            sims = corpus['indexes'][model][fn_vect(text, corpus)]
    return [s[0] for s in sorted(enumerate(sims), key=lambda item: -item[1])]
#     return [s[0] for s in sorted(sims, key=lambda item: -item[1])]

# obtains a query given corpus, query version and number
query_text = lambda c, n: c['queries'].iloc[n].txt


# obtains document rankings for a given corpus, query version and number, for a 
#  given model.
def rank(corpus, query_number, model, fn_vect, ntopics=None):
    return query_index(query_text(corpus, query_number), corpus, model, fn_vect, ntopics)

# obtains rankings for all models, given a query and a corpus.
def ranks(corpus, query_number, topn=20):
    ranks = [
        pd.DataFrame({'tfidf' : rank(corpus, query_number, 'tfidf', tfidf_vect)}),
        pd.DataFrame({'bm25' : rank(corpus, query_number, 'bm25', bm25_vect)}),
        pd.DataFrame({'lsi_%s'%t : rank(corpus, query_number, 'lsi', lsi_vect, t) for t in topics}),
        pd.DataFrame({'lda_%s'%t : rank(corpus, query_number, 'lda', lda_vect, t) for t in topics}),
    ]
    ranks = pd.concat(ranks, axis=1)
    if topn:
        return ranks.iloc[:topn]
    else:
        return ranks
```

```
corpus_a['ranks'] = [ranks(corpus_a, n) for n in range(30)]
corpus_b['ranks'] = [ranks(corpus_b, n) for n in range(30)]

corpus_a['ranks'][0]
```

```
def load_docs(corpus):
    docs = pd.read_csv(corpus['data_dir'] + '/docs.csv', index_col=0)
    docs['matter'] = docs.matter.apply(eval)
    docs['citations'] = docs.citations.apply(eval)
    corpus['docs'] = docs
load_docs(corpus_a)
load_docs(corpus_b)
corpus_a['docs']
```

```
import numpy as np
def dcg(gain_vector):
    discount = np.array([1] + [np.log2(i) for i in range(2, len(gain_vector) + 1)])
    return np.cumsum(gain_vector / discount)[-1]

def ideal_gain_vector(corpus, at=20):
    ig = pd.DataFrame({'q%s'%i: corpus['docs']['q%s'%i].sort_values(ascending=False).values for i in range(30)})
    ig = ig.head(at)
    return ig

corpus_a['ig'] = ideal_gain_vector(corpus_a)
corpus_b['ig'] = ideal_gain_vector(corpus_b)
corpus_a['ig']
```

```
def ideal_discounted_cumulative_gain(ig, at=20):
    idcg = pd.DataFrame({'q%s'%i: ig['q%s'%i].expanding().apply(dcg) for i in range(30)})
    idcg = idcg.head(at)
    return idcg    

corpus_a['idcg'] = ideal_discounted_cumulative_gain(corpus_a['ig'])
corpus_b['idcg'] = ideal_discounted_cumulative_gain(corpus_b['ig'])

corpus_a['idcg']
```

```
def gain_vector(corpus, query_num, model):
    return corpus['docs']['q%s'%query_num][corpus['ranks'][query_num][model]].values

def ndcg(corpus, model, at=19):
    model_g = pd.DataFrame({'q%s'%i: gain_vector(corpus, i, model) for i in range(30)})
    model_dcg = model_g.expanding().apply(dcg)
    model_ndcg = model_dcg / corpus['idcg']
    return model_ndcg.iloc[at]

corpus_a['ndcg'] = pd.DataFrame({r: ndcg(corpus_a, r) for r in corpus_a['ranks'][0].columns})
corpus_b['ndcg'] = pd.DataFrame({r: ndcg(corpus_b, r) for r in corpus_b['ranks'][0].columns})

corpus_a['ndcg']
```
